#### Semantic Search (Similarity Search)
1. Dot Product
2. Cosine Sim
3. Eculidean Distance(L2)

```python
import numpy as np

def dot_product(a, b):
    return np.dot(a, b)

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def euclidean_distance(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)

print("=== Cosine Similarity (Higher is better) ===")
print("Range: -1 to 1")

print("\n=== Dot Product (Higher is better) ===")
print("Range: -∞ to +∞ (depends on vector magnitude)")

print("\n=== Euclidean Distance (Lower is better) ===")
print("Range: 0 to +∞")
```

```python
query = "How to reduce heart disease risk?"

documents = [
    "Eating fiber reduces heart risk.",
    "Fruits and vegetables lower cardiovascular disease chances.",
    "Buying a new car improves driving comfort.",
    "Regular exercise improves heart health."
]
```

```python
google_embeddings_model = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")
```

```python
query_embeddings = google_embeddings_model.embed_query(query)
doc_embeddings = google_embeddings_model.embed_documents(documents)
```

```python
len(query_embeddings)
```

```python
len(doc_embeddings)
```

```python
result = []
```

```python
for doc, doc_emb in zip(documents, doc_embeddings):
    result.append({
        "query": query,
        "document": doc,
        "cosine":cosine_similarity(query_embeddings, doc_emb),
        "dot_product": dot_product(query_embeddings, doc_emb),
        "euclidean": euclidean_distance(query_embeddings, doc_emb)
    })
```

```python
result
```

#### Score = number of matching words

```python
def keyword_search(query, document):
    query_words = query.lower().replace("?","").split()
    result=[]

    for doc in document:
        doc_words = doc.lower().replace(".","").split()
        doc_count = Counter(doc_words)

        score = sum(doc_count[w] for w in query_words)
        result.append((doc, score))
    return sorted(result, key=lambda x: x[1], reverse=True)
```

```python
keyword_result = keyword_search(query, documents)
```
